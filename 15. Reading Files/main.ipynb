```python
FILE_NAME = "recipes.txt"


def read_file():
    try:
        with open(FILE_NAME, "r") as file:
            return file.read()
    except FileNotFoundError:
        return ""


def get_recipe_names():
    names = []
    for line in read_file().splitlines():
        if line.startswith("Recipe:"):
            names.append(line.replace("Recipe:", "").strip())
    return names


def view_recipe():
    recipes = get_recipe_names()

    if not recipes:
        print("\nNo recipes found.\n")
        return

    print("\nAvailable Recipes:")
    for i, name in enumerate(recipes, 1):
        print(f"{i}. {name}")

    choice = int(input("\nChoose recipe number: "))
    selected = recipes[choice - 1]

    content = read_file()
    blocks = content.split("\n\n")

    for block in blocks:
        if block.startswith(f"Recipe: {selected}"):
            print("\n" + block)
            return


def main():
    while True:
        print("\nüçΩÔ∏è Recipe Viewer App üçΩÔ∏è")
        print("1. View a recipe")
        print("2. Exit")

        choice = input("\nEnter choice: ")

        if choice == "1":
            view_recipe()
        elif choice == "2":
            print("\nGoodbye üëã")
            break
        else:
            print("\nInvalid choice!\n")


if __name__ == "__main__":
    main()
```
